## 3. 문자열 다루기 & 다양한 자료형의 사칙 연산

### 학습 목표

- 문자열의 기본 특성을 이해한다.
- 문자열 포매팅과 슬라이싱을 활용할 수 있다.
- 다양한 자료형으로 사칙 연산을 진행한다.

````python
# 문자열 기본
message = "Hello, Python!"
print(message)

# 문자열 포매팅
name = "LikeLion"
age = 25
print(f"{name}는 {age}살입니다.")

# 문자열 슬라이싱
print(message[0:5])  # 처음부터 5번째 문자까지
print(message[-7:])  # 뒤에서 7번째 문자부터 끝까지
````

````python
# 사칙 연산
a = 10
b = 3

print(f"a={a}, b={b}")
print(f"덧셈 연산: a+b={a + b}")
print(f"뺄셈 연산: a-b={a - b}")
print(f"곱셈 연산: a*b={a * b}")
print(f"나눗셈 연산: a/b={a / b}")
print(f"몫 연산: a//b={a // b}")
print(f"나머지 연산: a%b= {a % b}")
print(f"거듭제곱 연산: a**b= {a ** b}")
````

## 4. 자료구조

### 4.1 리스트 (List)

- 순서가 있는 변경 가능한 자료구조
- 소괄호 []로 생성

````python
# 리스트 생성
fruits = ["apple", "banana", "cherry"]
print(f"과일 목록: {fruits}")

# 리스트 요소 접근
print(f"첫 번째 과일: {fruits[0]}")

# 리스트 메소드
fruits.append("date")  # 요소 추가
print(f"과일 추가 후: {fruits}")

fruits.remove("banana")  # 요소 제거
print(f"바나나 제거 후: {fruits}")

print(f"과일 개수: {len(fruits)}")
````

## 5. 조건문

### 학습 목표

- if, elif, else 문을 이해하고 활용할 수 있다.
- 조건문을 사용하여 프로그램의 흐름을 제어할 수 있다.

````python
x = 10

if x > 0:
    print("x는 양수입니다.")
elif x < 0:
    print("x는 음수입니다.")
else:
    print("x는 0입니다.")

# 중첩 조건문
age = 20
has_id = True

if age >= 18:
    if has_id:
        print("술을 구매할 수 있습니다.")
    else:
        print("신분증이 필요합니다.")
else:
    print("미성년자는 술을 구매할 수 없습니다.")
````

## 6. 반복문

### 학습 목표

- for 문과 while 문의 차이를 이해하고 적절히 사용할 수 있다.
- 반복문을 사용하여 효율적으로 코드를 작성할 수 있다.

````python
# for 반복문
for i in range(5):
    print(f"반복 {i+1}")

# 리스트와 for 문
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# while 반복문
count = 0
while count < 3:
    print(f"카운트: {count}")
    count += 1

# break와 continue
for i in range(10):
    if i == 3:
        continue  # 3을 건너뛰고 계속
    if i == 7:
        break  # 7에서 반복 중단
    print(i)
````

### 실습: 반복문 활용하기

````python
# 1. 1부터 10까지의 숫자 중 짝수만 출력하는 프로그램을 for 문을 사용해 작성해보세요.
for i in range(1, 11):
    if i % 2 == 0:
        print(i)
````

````python
# 2. 사용자가 'q'를 입력할 때까지 계속해서 사용자의 입력을 받고 출력하는 프로그램을 while 문을 사용해 작성해보세요.
while True:
    user_input = input("문자열을 입력하세요 ('q'를 입력하면 종료): ")
    if user_input == 'q':
        break
    print(f"입력받은 문자열: {user_input}")
````

## 7. 함수

### 학습 목표

- 함수의 정의와 호출 방법을 이해한다.
- 매개변수와 반환값을 활용할 수 있다.

````python
# 기본 함수
def greet(name):
    return f"안녕하세요, {name}님!"

print(greet("Python"))

# 기본값 매개변수
def power(base, exponent=2):
    return base ** exponent

print(power(3))  # 3^2
print(power(3, 3))  # 3^3

# 가변 인자
def sum_all(*args):
    return sum(args)

print(sum_all(1, 2, 3, 4, 5))
````

### 실습: 함수 만들기

````python
# 1. 두 수를 인자로 받아 큰 수를 반환하는 함수를 만들어보세요.
def max_number(a, b):
    return a if a > b else b

print(f"3과 7 중 큰 수: {max_number(3, 7)}")
````

````python
# 2. 리스트를 인자로 받아 리스트의 모든 요소를 출력하는 함수를 만들어보세요.
def print_list(lst):
    for item in lst:
        print(item)

print_list(["apple", "banana", "cherry"])
````

## 8. 예외 처리

예외 처리를 통해 프로그램이 예상치 못한 상황에서도 정상적으로 동작하도록 할 수 있다.

````python
try:
    num = int(input("숫자를 입력하세요: "))
    result = 10 / num
    print(f"10 / {num} = {result}")
except ValueError:
    print("올바른 숫자를 입력해주세요.")
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
except Exception as e:
    print(f"예상치 못한 오류가 발생했습니다: {e}")
finally:
    print("예외 처리가 완료되었습니다.")
````

## 9. 간단한 프로젝트: 숫자 맞추기 게임

````python
import random

def guess_number():
    number = random.randint(1, 100)
    attempts = 0

    print("1부터 100 사이의 숫자를 맞춰보세요!")

    while True:
        try:
            guess = int(input("추측한 숫자를 입력하세요: "))
            attempts += 1

            if guess < number:
                print("더 큰 숫자입니다.")
            elif guess > number:
                print("더 작은 숫자입니다.")
            else:
                print(f"축하합니다! {attempts}번 만에 숫자를 맞추셨습니다.")
                break
        except ValueError:
            print("올바른 숫자를 입력해주세요.")

guess_number()
````
